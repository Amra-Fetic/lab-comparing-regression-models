10.Define a function that takes a pandas DataFrame as an input and returns two pandas DataFrames: the first containing numerical continuous columns and the second containing numerical discrete columns of the input dataframe. To this end, it might be helpful to count the number of unique values. The function must have an optional argument set by default to 36 to discriminate between continuous and discrete columns. Then, use it to create two new dataframes: continuous_df and discrete_df.


```python
def two_data_frames(dataframe, threshold=36):
    continuous_column = []
    discrete_column = []
    for column in dataframe.columns:
        if len(dataframe[column].unique()) > threshold:
            continuous_column.append(column)
        else:
            discrete_column.append(column)
    return continuous_column, discrete_column

# Use the function to create two new dataframes: continuous_df and discrete_df
continuous_column, discrete_column = two_data_frames(numerical_df)
continuous_df = numerical_df[continuous_column]
discrete_df = numerical_df[discrete_column]
```

Make sure to sort the bars by frequency ie: the most frequent values should be placed first. Hint: use .value_counts(). In addition, if the amount of unique values of a categorical column (cardinality) is six or more, the corresponding countplot should have the bars placed on the y-axis instead of the x-axis.

```python
categorical_df.head()
```

```python
def plot_categorical_countplots(categorical_df):
    # Iterate through columns
    for column in categorical_df.columns:
        # Check if the column is categorical
        if pd.api.types.is_categorical_dtype(categorical_df[column]) or pd.api.types.is_object_dtype(categorical_df[column]):
            # Check the cardinality of the column
            unique_values = categorical_df[column].nunique()
            
            # Set the orientation based on cardinality
            orient = 'v' if unique_values >= 6 else 'h'
            
            # Create a countplot
            plt.figure(figsize=(10, 6))
            sns.countplot(data=categorical_df, x=column, order= categorical_df[column].value_counts().index, orient=orient)
            
            # Set plot labels and title
            plt.xlabel(column)
            plt.ylabel('Count')
            plt.title(f'Countplot of {column}')
            
            # Rotate x-axis labels if horizontal orientation
            if orient == 'h':
                plt.xticks(rotation=45)
                plt.show()
                
plot_categorical_countplots(categorical_df)
```

13.Define a function to remove the outliers of a numerical continuous column depending if a value is bigger or smaller than a given amount of standard deviations of the mean (thr=3).


```python
# Creating a function to remove outliers
def remove_outliers(df, columns, threshold=3):
    """
    Remove outliers from specified numerical columns in a DataFrame.

    Parameters:
    - df (pd.DataFrame): The DataFrame containing the data.
    - columns (list): List of column names for which outliers will be removed.
    - threshold (int or float): The threshold in terms of standard deviations from the mean.
    
    Returns:
    - pd.DataFrame: The DataFrame with outliers removed.
    """

    df_no_outliers = df.copy()

    for column_name in columns:
        # Calculate mean and standard deviation
        mean_value = df[column_name].mean()
        std_dev = df[column_name].std()

        # Define the lower and upper bounds for outliers
        lower_bound = mean_value - threshold * std_dev
        upper_bound = mean_value + threshold * std_dev

        # Remove outliers for the current column
        df_no_outliers = df_no_outliers[(df_no_outliers[column_name] >= lower_bound) & (df_no_outliers[column_name] <= upper_bound)]

    return df_no_outliers


```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
def calculate_error_metrics(y_true, y_pred):
    # Calculate error metrics
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mape = mean_absolute_percentage_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    # Create a Pandas DataFrame
    metrics_df = pd.DataFrame({
        'Error Metric': ['MAE', 'MSE', 'RMSE', 'MAPE', 'R2'],
        'Value': [mae, mse, rmse, mape, r2]
    })

    return metrics_df
def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

```
